```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("ALL.csv", parse_dates=["Date_time"])
df.head(5)
```

```python
new_index = pd.date_range(start = df["Date_time"].min(), end = df["Date_time"].max(), freq = "5T" )
new_df = pd.DataFrame(new_index, columns=['Date_time'])
new_df.head(5)
```

```python
# Kết hợp DataFrame mới với DataFrame gốc
merged_df = pd.merge(new_df, df, how='left', on='Date_time')
feature = merged_df.drop("Date_time", axis = 1).columns
# Nội suy tuyến tính trên các khoảng thời gian cụ thể
merged_df[feature] = merged_df[feature].interpolate(method='linear')
# merged_df[feature] = merged_df[feature].interpolate(method='linear interp')
merged_df.to_csv("ALL_5T.csv")
print(merged_df.columns)
```
